```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import sklearn
import statistics as st
from datetime import datetime
import matplotlib.dates as mdates
import time
```

```python
#import data
stockData = 'AAPL.csv'
data = pd.read_csv(stockData)
```

```python
## Simple moving average attempt
## Formula: SMA = (P1 +P2 +P3 +...Pn)/n
# first predicted value = start + window
vals = 50
window = 5
prediction = np.zeros(vals)
sum = 0
start = 45
for j in range(vals):
    sum = 0
    for i in range(start,start+window):
        sum += data['High'][i]
    prediction[j] = sum/window
    start+=1
    #print(sum/n)
```

```python
###plotting data
x = np.arange(100)
plt.plot(x,data['High'][0:100],label= 'true')
plt.plot(x[50:100],prediction,label = 'prediction')
plt.legend()
```
