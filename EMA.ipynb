```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import sklearn
import statistics as st
from datetime import datetime
import matplotlib.dates as mdates
import time
```

```python
#import data 
stockData = 'AAPL.csv'
data = pd.read_csv(stockData)
```

```python
## Exponential moving average attempt
## Formula: EMA = Pt * k + EMA(t-1) *(1-k)
## k = 2/(N+1)
# first predicted value = start + window

# Calculate SMA of first 5 numbers
EMA = []
vals = 1
window = 5
first5 = 0
sum = 0
start = 45
current = start
for i in range(start,start+window):
    sum += data['High'][i]
first5 = sum/window
current += window
EMA.append(first5)

#calculate EMA values
N = window
k = 2/(N+1)
index = current
for i in range(1,50):
    value = data['High'][index] *k + EMA[i-1] * (1-k)
    EMA.append(value)
    index +=1
```

```python
###plotting data
x = np.arange(100)
plt.plot(x,data['High'][0:100],label= 'true')
plt.plot(x[50:100],EMA,label = 'prediction')
plt.legend()
```
